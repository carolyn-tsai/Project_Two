```python
# Import dependencies

import os
import csv
import pandas as pd

file = "wheelchair_data_w_address.csv"

wc_df = pd.read_csv(file, encoding="UTF-8", low_memory = False)    

# Preview Data Frame

wc_df.head()
```

```python
# Check number of rows

len(wc_df)
```

```python
# Check if any of the restaurants are closed
# If there are any rows with "True", drop those rows

wc_df[wc_df['is_closed'] == "True"]

# No entries with "True"
```

```python
# Drop any NaN entries in data

no_nan = wc_df.dropna()

# See if any rows were dropped

len(no_nan)
```

```python
# Create new data frame with specific columns

wc_data = no_nan[["name", "latitude", "longitude", "address", "zip_code", "url"]]

# Preview data frame

wc_data.head()
```

```python
# Test to see if address entry could be cleaned up
# Address format is currently [street, city, state/zip] but needs to be as one entry

wc_data["address"][1][2:-2]
```

```python
# Create new column "address_clean" and set it equal to sliced address values 

wc_data["address_clean"] = wc_data["address"].str.slice(2,-2)

# Preview the data frame

wc_data.head()
```

```python
# Check to see if the slice produced any NaN values
# Drop any NaN values

check_nan = wc_data.dropna()

# Check to see if any rows were dropped

len(check_nan)
```

```python
# Check one of the entries to see what format is

wc_data["address_clean"][10]
```

```python
# Replace "'" with "" to get rid of extra characters

wc_data["address_clean"][10].replace("'", "")
```

```python
# Replace all address column entries with correctly formatted address

wc_data["address"] = wc_data["address_clean"].str.replace("'", "")

# Preview data frame

wc_data.head()
```

```python
# Get final data frame with correct columns for wheel chair data
# Will need to compare this data with the dog friendly restaurant data

final_wc_df = wc_data[["name", "latitude", "longitude", "address", "zip_code", "url"]]

# Preview data frame

final_wc_df.head()
```

```python
# Rename the "url" column to "website" to match the dog friendly restaurant data

final_wc_df = final_wc_df.rename(columns={"url":"website"})

# Preview data frame

final_wc_df.head()
```

```python
# Drop duplicate entries, if any
# Look by "name", "latitude", and "longitude" column as some restaurants are chains and will have same name but different locations

final_wc_df[final_wc_df[["name", "latitude", "longitude"]].duplicated() == True]
```

```python
# Check number of rows of data frame

len(final_wc_df)
```

```python
# Save data frame as CSV file

final_wc_df.to_csv("final_wheelchair_data.csv", index = False)
```
