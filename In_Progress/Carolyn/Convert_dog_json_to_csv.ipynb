```python
# Import dependencies 

import json
import pandas as pd
from pandas.io.json import json_normalize

# Load the dog json file and set it to a variable to work with

with open("dog_sf_data.json", encoding="utf8") as json_file:
    data_set = json.load(json_file)
```

```python
# Turn the json file into a pandas dataframe to view and analyze data

dog_df = pd.DataFrame.from_dict(json_normalize(data_set), orient="columns")
```

```python
# Preview data frame to make sure everything is in the right column

dog_df.head()
```

```python
# Get rid of unnecessary columns by selecting specific columns for the dataframe

clean_dog_df = dog_df[["website", "name", "result_type", "longitude", "latitude", "address"]]

# Display the new dataframe

clean_dog_df.head()
```

```python
# Check how many rows are in dataframe

len(clean_dog_df)
```

```python
# Get rid of any NaN values

clean_dog_df = clean_dog_df.dropna()
```

```python
# Check updated number of rows

len(clean_dog_df)
```

```python
# Save the dataframe as CSV

clean_dog_df.to_csv("clean_sf_dog_data.csv", index=False)
```
